```python
# chia ra 5 nhóm tuổi
df_bank_processed.loc[(df_bank_processed['age'] > 15) & (df_bank_processed['age'] <= 30), 'age'] = 1
df_bank_processed.loc[(df_bank_processed['age'] > 30) & (df_bank_processed['age'] <= 40), 'age'] = 2
df_bank_processed.loc[(df_bank_processed['age'] > 40) & (df_bank_processed['age'] <= 50), 'age'] = 3
df_bank_processed.loc[(df_bank_processed['age'] > 50) & (df_bank_processed['age'] <= 60), 'age'] = 4
df_bank_processed.loc[ df_bank_processed['age'] > 60, 'age'] = 5
```

### 1. What profiles ( age / job / marital status / etc.) are the most likely to respond positively ?

```python
grouped_data = df_bank_processed.groupby('y')

def find_max_response_value(series):
    response_rates = series.value_counts(normalize=True)
    max_response_value = response_rates.idxmax()
    return max_response_value

max_response_age = grouped_data['age'].apply(find_max_response_value).loc['yes']
print("Độ tuổi gửi tiền vào ngân hàng nhiều nhất thuộc nhóm:", max_response_age)
max_response_job = grouped_data['job'].apply(find_max_response_value).loc['yes']
print("công việc gửi tiền vào ngân hàng nhiều nhất là:", max_response_job)
max_response_marital = grouped_data['marital'].apply(find_max_response_value).loc['yes']
print("người có tình trạng hôn nhân gửi tiền vào ngân hàng nhiều nhất thuộc nhóm:", max_response_marital)
max_response_education = grouped_data['education'].apply(find_max_response_value).loc['yes']
print("người có tình trạng giáo dục gửi tiền vào ngân hàng nhiều nhất thuộc nhóm:", max_response_education)
```

### 2. Does the financial situation of people have an impact on how they answer ?

```python
# Chuyển đổi giá trị 'yes' thành 1 và 'no' thành 0 trong cột 'Response'
df_bank_processed['y'] = df_bank_processed['y'].map({'yes': 1, 'no': 0})
feedback_by_default = df_bank_processed.groupby('default')['y'].mean()
feedback_by_housing = df_bank_processed.groupby('housing')['y'].mean()
feedback_by_loan = df_bank_processed.groupby('loan')['y'].mean()

print("Tỷ lệ gửi tiền theo cột 'default':")
print(feedback_by_default.round(3))

print("Tỷ lệ gửi tiền theo cột 'housing':")
print(feedback_by_housing.round(3))

print("Tỷ lệ gửi tiền theo cột 'loan':")
print(feedback_by_loan.round(3))
```

### 3. Do the timing (date / duration) and type of contact matter significantly ?

```python

# Tạo danh sách các nhóm bạn muốn phân tích
time_month = ['month']
time_day_of_week = [ 'day_of_week']
contact_groups = ['contact']


response_rates_time = {}
response_rates_duration = {}
response_rates_contact = {}

# Tính tỷ lệ phản hồi cho mỗi nhóm thời gian
for group in time_groups:
    grouped_data = df_bank_processed.groupby(group)
    response_rates_time[group] = grouped_data['y'].value_counts(normalize=True).unstack().fillna(0)


# Tính tỷ lệ phản hồi cho mỗi loại liên lạc (contact)
for group in contact_groups:
    grouped_data = df_bank_processed.groupby(group)
    response_rates_contact[group] = grouped_data['y'].value_counts(normalize=True).unstack().fillna(0)

# In kết quả
for group, rates in response_rates_time.items():
    print(f'Tỷ lệ phản hồi theo {group}:')
    print(rates.round(3))

for group, rates in response_rates_duration.items():
    print(f'Tỷ lệ phản hồi theo {group}:')
    print(rates.round(3))

for group, rates in response_rates_contact.items():
    print(f'Tỷ lệ phản hồi theo {group}:')
    print(rates.round(3))
```

### 4 - What is the best number of contacts to reach agreement ?
### What is the best delay ?
### Are people more likely to agree to the campaign if they have already agreed in the past ?

```python

best_campaign = df_bank_processed[df_bank_processed['y'] == 1]['campaign'].idxmin()
best_campaign_value = df_bank_processed.loc[best_campaign, 'campaign']

agreed_before = df_bank_processed[df_bank_processed['previous'] > 0]
agreed_before_rate = agreed_before['y'].value_counts(normalize=True).get(1, 0)

#Lọc ra các dòng có pdays khác -1 (đã được liên hệ trước đó)
filtered_data = df_bank_additional[df_bank_additional['pdays'] != -1]
average_delay_to_agree = filtered_data.groupby('pdays')['y'].mean()
best_delay = average_delay_to_agree.idxmax()

print("Khoảng thời gian tối ưu giữa các cuộc gọi tiếp thị:",best_delay, "ngày")

print(f"Số lần liên hệ tốt nhất để có thỏa thuận: {best_campaign_value}")

print(f"Tỷ lệ đạt thỏa thuận khi đã có thỏa thuận trước đây: {agreed_before_rate:.2%}")
```

### 5. Does the economy (Employment variation rate / Consumer Price Index  / Consumer Confidence Index / Euribor / etc.) have an impact ? How ? Is it the same for everybody ?

```python
feedback_by_emp = df_bank_additional.groupby('emp.var.rate')['y'].mean()

print("\nTỷ lệ gửi tiền theo Employment variation rate")
print(feedback_by_emp)

feedback_by_cons_price = df_bank_additional.groupby('cons.price.idx')['y'].mean()

print("\nTỷ lệ gửi tiền theo Consumer Price Index")
print(feedback_by_cons_price)

feedback_by_cons_conf = df_bank_additional.groupby('cons.conf.idx')['y'].mean()

print("\nTỷ lệ gửi tiền theo Consumer Price Index")
print(feedback_by_cons_conf)
```

#### Nhận xét
1. Tỷ lệ gửi tiền theo Employment Variation Rate (emp.var.rate):

    - Có một biểu đồ tỷ lệ gửi tiền đối với Employment Variation Rate. Tỷ lệ phản hồi cao nhất xảy ra khi Employment Variation Rate ở mức -3.0 và -1.7, với tỷ lệ phản hồi lần lượt là 51.16% và 52.13%. Điều này ngụ ý rằng khi tỷ lệ biến đổi việc làm (Employment Variation Rate) thấp hoặc âm (có giảm sút), tỷ lệ gửi tiền cao hơn.

    - Khi Employment Variation Rate tăng lên (ví dụ: 1.4), tỷ lệ gửi tiền giảm xuống 5.33%. Điều này cho thấy một sự tương quan ngược giữa Employment Variation Rate dương và tỷ lệ gửi tiền.

2. Tỷ lệ gửi tiền theo Consumer Price Index (cons.price.idx):

    - Tỷ lệ gửi tiền dao động dọc theo Consumer Price Index. Tỷ lệ gửi tiền cao nhất xuất hiện khi Consumer Price Index nằm trong khoảng từ 93.369 đến 93.798, với tỷ lệ phản hồi từ 55.74% đến 62.69%. Điều này ngụ ý rằng khi chỉ số giá tiêu dùng (Consumer Price Index) ở mức này, tỷ lệ gửi tiền là cao nhất.

    - Khi Consumer Price Index tăng hoặc giảm khá xa khỏi khoảng giá trị này, tỷ lệ gửi tiền giảm xuống (ví dụ: 92.201 hoặc 94.027), chỉ còn khoảng 3-5%. Điều này cho thấy rằng có một mối quan hệ tương quan giữa Consumer Price Index và tỷ lệ gửi tiền.

3. Tỷ lệ gửi tiền theo Consumer Confidence Index (cons.conf.idx):

    - Tỷ lệ gửi tiền cũng thay đổi dựa trên Consumer Confidence Index. Tỷ lệ gửi tiền cao nhất xuất hiện khi Consumer Confidence Index nằm trong khoảng từ -40.0 đến -38.3, với tỷ lệ phản hồi từ 51.50% đến 62.69%. Điều này ngụ ý rằng khi chỉ số niềm tin của người tiêu dùng (Consumer Confidence Index) ở mức này, tỷ lệ gửi tiền là cao nhất.

    - Khi Consumer Confidence Index giảm thấp hơn hoặc tăng cao hơn, tỷ lệ gửi tiền giảm xuống (ví dụ: -36.4 hoặc -26.9), chỉ còn khoảng 3-5%. Điều này cho thấy rằng có một mối quan hệ tương quan giữa Consumer Confidence Index và tỷ lệ gửi tiền.

### 6. Any interesting insights to add ?

```python
# Tạo một DataFrame để lưu trữ thông kê chung
summary_data = pd.DataFrame(columns=['Column', 'Top Response', 'Response Rate'])

# Lặp qua từng cột trong DataFrame và xác định giá trị nổi trội có tỷ lệ phản hồi cao nhất
for column in df_bank_additional.columns:
    if df_bank_additional[column].dtype == 'object':  # Chỉ xem xét các cột có kiểu dữ liệu là object (categorical)
        top_response = df_bank_additional[column].value_counts().idxmax()
        response_rate = df_bank_additional[df_bank_additional[column] == top_response]['y'].mean()
        summary_data = pd.concat([summary_data, pd.DataFrame({'Column': [column], 'Top Response': [top_response], 'Response Rate': [response_rate]})], ignore_index=True)

# Hiển thị thông kê chung
print(summary_data)
```
